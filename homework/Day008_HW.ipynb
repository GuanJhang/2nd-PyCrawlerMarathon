### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
r.encoding = 'utf-8'

response = r.text

soup = BeautifulSoup(response,'html.parser')
print(soup)
#print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')
print("資料型態為html")
print(".text 可以將得到的資料已string方式儲存")
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')
print("以BeautifulSoup處理的好處是，它可以將string轉換成List方便擷取")
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
response = r.text
soup = BeautifulSoup(response,'html.parser')
print(soup)
print(r.text[0:600])
```
