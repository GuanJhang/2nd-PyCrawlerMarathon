## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    time = d.find(class_ = "date").text[-5:]
    break
timesec = int(time[0:2])*3600 + int(time[3:])*60

for d in soup.find(class_="part_list_2").find_all('h3'):
    if(timesec - 7200 <= int(d.find(class_ = "date").text[-5:][0:2])*3600 + int(d.find(class_ = "date").text[-5:][3:])*60):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

```

```python
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2020')
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('3')
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value('23')

browser.find_element_by_id('button').click()
```

```python
# 每個兩秒鐘自動往下滑

import time
for i in range(100):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 0000);")
```
