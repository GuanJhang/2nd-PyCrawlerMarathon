### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
flag = 0
cat = []
cate = []


for d in soup.find(class_ = "part_list_2").find_all('h3'):
    cate.append(d.find_all('em')[-1].text)
    # print(d.find_all('em')[-1].text)
cate = sorted(cate)
# print(cate)
cat.append(cate[0])

for word in cate:
    if(cat[flag] == word):
        continue
    else:
        flag += 1
        cat.append(word)
        
print(cat)       # cat 為所有分類
```

### ③ 哪一個類別的文章最多
