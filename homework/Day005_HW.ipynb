```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)

count = 0
for d in data:
    count += 1
string = "這個 API 一次會回傳 " + str(count) + " 筆資料\n"
print(string)

print("包含這些資料:")
for d in data[0]:
    print(d)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data:
    print("標題:")
    print(d['title'])
    print("貼文時間")
    print(d['createdAt'])
    print("留言人數")
    print(d['commentCount'])
    print("按讚人數")
    print(d['likeCount'])
    print("\n")
```
